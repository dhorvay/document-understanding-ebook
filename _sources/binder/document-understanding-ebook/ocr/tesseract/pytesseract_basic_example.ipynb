This notebook covers the most basic pytesseract use case - extracting text from images.

## Load image using Pillow

pytesseract has good support for [Pillow](https://pypi.org/project/Pillow/), so we will use that to load our images.
Here we will be using a [simple letter-style document](https://commons.wikimedia.org/wiki/File:Bill_Gates_Letter_to_Hobbyists.jpg), which gets decent quality results out of the box.

```python
path = '../../../binder-datasets/ocr/images/letter.jpg'
im = Image.open(path)
display(im.resize(int(0.25*s) for s in im.size))
```

## Run pytesseract and extract text from an image

Let's run the image through pytesseract and print the output text.

```python
output = pytesseract.image_to_string(im)
print(output)
```
